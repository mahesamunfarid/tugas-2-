```python
from matplotlib import pyplot as plt

x = [10, 30, 50]

y0 = [20, 30, 40]
y = [25, 15, 30]


plt.plot(x, y, color="green")
plt.plot(x, y0)

plt.title("Contoh plot")
plt.xlabel("nilai x")
plt.ylabel("nilai y")
plt.show()

data_rekap = [
    {
        "tahun": 1995,
        "penjualan": 300,
        "operasional": 150,
        "pajak": 50,
        "pendapatan_lain": 135
    },
    {
        "tahun": 1997,
        "penjualan": 320,
        "operasional": 120,
        "pajak": 70,
        "pendapatan_lain": 105
    },
    {
        "tahun": 1991,
        "penjualan": 200,
        "operasional": 75,
        "pajak": 35,
        "pendapatan_lain": 85
    },
    {
        "tahun": 1992,
        "penjualan": 320,
        "operasional": 170,
        "pajak": 80,
        "pendapatan_lain": 235
    },
    {
        "tahun": 2001,
        "penjualan": 300,
        "operasional": 150,
        "pajak": 50,
        "pendapatan_lain": 135
    },
    {
        "tahun": 2003,
        "penjualan": 500,
        "operasional": 350,
        "pajak": 100,
        "pendapatan_lain": 55
    },
    {
        "tahun": 2005,
        "penjualan": 750,
        "operasional": 450,
        "pajak": 200,
        "pendapatan_lain": 155
    }
]
# expected data
data_graph = {
    "tahun": [...], # x
    "penjualan": [...] # y
}
# mapping
# data_map = { <tahun>: <penjualan>}
data_map = { data["tahun"]: data["penjualan"] for data in data_rekap}
data_map
{1995: 300, 1997: 320, 1991: 200, 1992: 320, 2001: 300, 2003: 500, 2005: 750}
# Sorting data tahun
tahun = list(data_map.keys())
tahun.sort()
tahun
[1991, 1992, 1995, 1997, 2001, 2003, 2005]
# get data penjualan
penjualan = [data_map[_tahun]for _tahun in tahun]
penjualan
    
[200, 320, 300, 320, 300, 500, 750]
data_graph = {
    "tahun": tahun,
    "penjualan": penjualan,
    "pajak": [20, 30, 30, 20, 30, 50, 75]
}
data_graph
{'tahun': [1991, 1992, 1995, 1997, 2001, 2003, 2005],
 'penjualan': [200, 320, 300, 320, 300, 500, 750],
 'pajak': [20, 30, 30, 20, 30, 50, 75]}
plt.plot(data_graph["tahun"], data_graph["penjualan"],"o-", label="penjualan")
plt.plot(data_graph["tahun"], data_graph["pajak"],"o-", label="pajak")

plt.title("Data rekapitulasi Tahunan")
plt.legend(loc="upper left")
plt.xlabel("Tahun")
plt.ylabel("Dalam jutaan Rp.")
plt.show()

import matplotlib.pyplot as plt

# Data rekap yang ada
data_rekap = [
    {"tahun": 1995, "penjualan": 300, "operasional": 150, "pajak": 50, "pendapatan_lain": 135},
    {"tahun": 1997, "penjualan": 320, "operasional": 120, "pajak": 70, "pendapatan_lain": 105},
    {"tahun": 1991, "penjualan": 200, "operasional": 75, "pajak": 35, "pendapatan_lain": 85},
    {"tahun": 1992, "penjualan": 320, "operasional": 170, "pajak": 80, "pendapatan_lain": 235},
    {"tahun": 2001, "penjualan": 300, "operasional": 150, "pajak": 50, "pendapatan_lain": 135},
    {"tahun": 2003, "penjualan": 500, "operasional": 350, "pajak": 100, "pendapatan_lain": 55},
    {"tahun": 2005, "penjualan": 750, "operasional": 450, "pajak": 200, "pendapatan_lain": 155}
]

# Mendapatkan data untuk tahun, penjualan, pajak, operasional, dan pendapatan lain
tahun = [data["tahun"] for data in data_rekap]
penjualan = [data["penjualan"] for data in data_rekap]
pajak = [data["pajak"] for data in data_rekap]
operasional = [data["operasional"] for data in data_rekap]
pendapatan_lain = [data["pendapatan_lain"] for data in data_rekap]

# Menghitung laba: laba = penjualan + pendapatan_lain - pajak - operasional
laba = [penj + pend_lain - pjk - opr for penj, pend_lain, pjk, opr in zip(penjualan, pendapatan_lain, pajak, operasional)]

# Menggabungkan data ke dalam dictionary
data_graph = {
    "tahun": tahun,
    "penjualan": penjualan,
    "pajak": pajak,
    "operasional": operasional,
    "pendapatan_lain": pendapatan_lain,
    "laba": laba
}

# Plot beberapa variabel dari data
plt.plot(data_graph["tahun"], data_graph["penjualan"], "o-", label="penjualan")
plt.plot(data_graph["tahun"], data_graph["pajak"], "o-", label="pajak")
plt.plot(data_graph["tahun"], data_graph["operasional"], "o-", label="operasional")
plt.plot(data_graph["tahun"], data_graph["pendapatan_lain"], "o-", label="pendapatan lain")
plt.plot(data_graph["tahun"], data_graph["laba"], "o-", label="laba")

# Menambahkan label dan judul
plt.title("Data Rekapitulasi Tahunan")
plt.xlabel("Tahun")
plt.ylabel("Jumlah dalam Jutaan Rp")
plt.legend(loc="upper left")

# Menampilkan grafik
plt.show()
```
